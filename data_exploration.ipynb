## Data Understanding and preparation

```python
data.describe(include=[object])
```

Conclusions from the chosen dataset description:
1) Education and Marital_Status have a total of 5 and 8 unique values, out of 2240 count. That means that these are categorical variables.

```python
data.describe()
```

```python
data["Age"] = 2023 - data["Year_Birth"]
data["Kids"] = data["Kidhome"] + data["Teenhome"]
data["Is_Parent"] = data["Kids"].apply(lambda kids: 1 if kids > 0 else 0)
data["Spent"] = data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]
data["Marital_Status"]=data["Marital_Status"].replace({"Married":"Together", "Together":"Together","Absurd":"Single", "Widow":"Single", "YOLO":"Single", "Divorced":"Single", "Single":"Single", "Alone":"Single"})
data["Education"]=data["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold","NumDealsPurchases": "DealsPurch", "NumWebPurchases": "WebPurch", "NumCatalogPurchases" : "CatalogPurch", "NumStorePurchases": "StorePurch", "NumWebVisitsMonth": "WebVisits"})
to_drop = ["Year_Birth","Z_CostContact", "Z_Revenue","Kidhome","Teenhome"]
data = data.drop(to_drop, axis=1)
```

By analysing our numerical variables we decided to make some changes to better analyse the data:
1) We created an "Age" column to indicate the age of the costumer. We did that using the "Year_Birth" column that holds the birth years of every costumer;
2) We created an "Kids" column to indicate the total of children in a household. We did that by adding the "Kidhom" and "Teenhome" columns;
3) We created a feature "Is_Parent" to indicate the parenthood status of each person;
4) We created an "Spent" column to indicate the total amount spent by the costumer in various categories. We did that by adding the columns that correspond to the products.
5) We reduced the number of categories in the 'Marital_Status' column from eight to two, and in the 'Education' column from five to three, to simplify.
6) We renamed some of the columns to more simple and intuitive names.
7) We dropped some of the redundant features.

## Visual analysis

### Distribution of the categorical variables

```python
cols = ["Education", "Marital_Status"]
fig_nr_columns = 2
fig_nr_rows = 1 + (len(cols) // fig_nr_columns) if len(cols) % fig_nr_columns != 0 else len(cols) // fig_nr_columns
fig = plt.figure(1, figsize=[20,5])

for k in range(len(cols)):
    data_agg = data.groupby(cols[k]).size().reset_index()
    x = data_agg[cols[k]]
    y = data_agg[0]
    ypos = np.arange(len(x))
    ax = fig.add_subplot(fig_nr_rows, fig_nr_columns, k+1)
    plt.subplots_adjust(hspace=0.3, wspace=0.2)
    ax.bar(ypos, y, color='purple')
    ax.tick_params(axis='x', rotation=45)
    plt.xticks(ypos, x)
    ax.set_title("Column: {}".format(cols[k]), fontweight='bold', fontsize=16)  # Adjust the fontsize as needed
    ax.set_ylabel("Count", fontsize=12)  # Adjust the fontsize as needed

plt.show()
```

### Multivariate anlysis

We select some features to see the relation between them, to see the broader view of the data.

```python
sns.set(rc={"axes.facecolor":"#FFF9ED","figure.facecolor":"#FFF9ED"})
pallet = ["#6a0d83", "#5B267B", "#726974", "#42838C", "#119DA4", "#19647E"]
cmap = colors.ListedColormap(["#6a0d83", "#5B267B", "#726974", "#42838C", "#119DA4", "#19647E"])
To_Plot = [ "Income", "Recency", "WebVisits", "Age", "Spent","Is_Parent"]
plt.figure()
sns.pairplot(data[To_Plot], hue= "Is_Parent",palette= (["#6a0d83", "#19647E" ]))
plt.show()
```

With this plot we form some conclusions:

1) Individuals with children tend to have significantly higher incomes compared to those without kids.
2) Those who spend more on shopping are typically individuals without children, and they also tend to be younger.
3) Individuals with kids tend to make more web visits compared to those without children.
4) Parents exhibit a greater time gap between purchases, indicating a more deliberate and considered approach to buying.

In summary, individuals who are parents appear to demonstrate greater financial responsibility and purchasing prudence compared to those without children.

### Numerical vs categorical variables

```python
import matplotlib.pyplot as plt
import seaborn as sns

variables = ["Spent", "Income", "Recency", "Kids", "WebVisits"]
titles = ["Spent vs Education", "Income vs Education", "Recency vs Education", "Kids vs Education", "WebVisits vs Education"]

plt.figure(figsize=(10,2))

for i, variable in enumerate(variables):
    sns.catplot(x="Education", y=variable, data=data, kind="box", aspect=1.5, height=4)
    plt.title(titles[i])

plt.tight_layout()
plt.show()
```

In this first set of plots we compared the categorical variable 'Education' with some numerical variables, and concluded the following:

- Spent: The plot for 'Spent' indicates that, on average, individuals with an undergraduate education tend to spend less in the overall context;

- Income: The 'Income' graph reveals that individuals with an undergraduate education generally earn less compared to those with higher education levels;

- Recency: The results for 'Recency' appear to be relatively consistent across different education levels;

- Kids: The 'Kids' graph suggests similar outcomes across various education levels;

- WebVisits: The 'WebVisits' graph illustrates that individuals with an undergraduate education tend to engage in more web visits.
